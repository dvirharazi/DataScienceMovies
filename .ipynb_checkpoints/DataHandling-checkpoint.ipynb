```python
import numpy as np
import pandas as pd
import matplotlib as plot
```

# Data Handling

```python
def clean_dataframe(df):

    df = df.drop(df.columns[0], axis=1)
    df = df.dropna(axis=0, subset=['star_1','star_2','star_3'])
    df = df.dropna(subset=['Length Of Movie (minute)'])
    df = df.dropna(subset=['idbm_year'])
    df["idbm_year"] = df["idbm_year"].astype(int)
    df = df[df.year == df.idbm_year]
    df = df.reset_index(drop=True)

    return df
```

In this function, we reformat the length of the movie - from an hour and min to min

```python
def handling_length_of_movie(df):
    length_in_minute = []
    length_col = df["length"].fillna("")
    for row in range(length_col.shape[0]):
        if any(char.isdigit() for char in length_col[row]) and 'TV' not in length_col[row] and 'PG' not in length_col[row]:
            time_minute = length_col[row].split(" ")
            if len(time_minute) == 2:
                hour = list(time_minute[0])
                hour_dig = list(hour)
                h = int(''.join(hour_dig[:-1])) * 60
                
                minute = list(time_minute[1])
                minute_dig = list(minute)
                m = int(''.join(minute_dig[:-1]))
                
                length_in_minute.append(int(m + h))
            elif len(time_minute) == 1:
                time = list(time_minute[0])
                time_dig = list(time)
                if hour_dig[-1] == 'h':
                    t = int(''.join(hour_dig[:-1]))
                else:
                    t = int(''.join(hour_dig[:-1])) * 60
                length_in_minute.append(t) 
        else:
            length_in_minute.append(np.nan)
            
    df.insert(2, "Length Of Movie (minute)", length_in_minute, True)
    df = df.drop('length', 1)
    return df
                

                
    
    
```

In this function we change a table to numeric - every genre will have a number, we do that for the ML.

```python
def change_to_numeric(df, col_to_change):  ##[genre_1, genre_2,genre_3]
    values_dict = dict()
    values_dict[np.nan] = 0
    value = 1
    number_of_col = len(col_to_change)
    for i in range(number_of_col):
        genres = df[col_to_change[i]]
        for genre in genres:
            if genre not in values_dict:
                values_dict[genre] = value
                value += 1
                
    for i in range(number_of_col):
        df = df.replace({col_to_change[i]:values_dict})
    return df
```

```python
df = pd.read_csv('movie_wiki_done.csv')
df = handling_length_of_movie(df)
df

# df.describe(include='all')



```
