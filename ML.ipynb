```python
# imports 
import pandas as pd
import numpy as np
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler, StandardScaler, normalize
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, f1_score
```

```python
df = pd.read_csv('movie_table_finish.csv') #load data


df = df.drop(['Unnamed: 0'], axis=1).copy()
# df['rating'] = df['rating'].apply(np.ceil)
df
```

```python
df.describe()
```

```python

df['director movie amount'] = df.groupby('director_1')['director_1'].transform('count')
df['writer1 movie amount'] = df.groupby('writer_1')['writer_1'].transform('count')
df['star1 movie amount'] = df.groupby('star_1')['star_1'].transform('count')
df['star2 movie amount'] = df.groupby('star_2')['star_2'].transform('count')
df['star3 movie amount'] = df.groupby('star_3')['star_3'].transform('count')

df['average rating genre'] = round(df.groupby('genre_1')['rating'].transform('mean'),3)
df['director rating avg'] = round(df.groupby('director_1')['rating'].transform('mean'),3)
df['star1 rating avg'] = round(df.groupby('star_1')['rating'].transform('mean'),3)
df['star2 rating avg'] = round(df.groupby('star_2')['rating'].transform('mean'),3)
df['star3 rating avg'] = round(df.groupby('star_3')['rating'].transform('mean'),3)
df['writer1 rating avg'] = round(df.groupby('writer_1')['rating'].transform('mean'),3)

df['Action'] = df[df['genre_1']=='Action'].groupby('director_1')['name'].transform('count')
df['Drama'] = df[df['genre_1']=='Drama'].groupby('director_1')['name'].transform('count')
df['Comedy'] = df[df['genre_1']=='Comedy'].groupby('director_1')['name'].transform('count')

df['Comedy'] = df['Comedy'].fillna(0)
df['Action'] = df['Action'].fillna(0)
df['Drama'] = df['Drama'].fillna(0)
df['average rating genre'] = df['average rating genre'].fillna(0)
df['director rating avg'] = df['director rating avg'].fillna(0)
df['star1 rating avg'] = df['star1 rating avg'].fillna(0)
df['star2 rating avg'] = df['star2 rating avg'].fillna(0)
df['star3 rating avg'] =df['star3 rating avg'].fillna(0)
df['writer1 rating avg'] = df['writer1 rating avg'].fillna(0)
df['director movie amount'] =df['director movie amount'].fillna(0)
df['writer1 movie amount'] = df['writer1 movie amount'].fillna(0)
df['star1 movie amount'] =df['star1 movie amount'].fillna(0)
df['star2 movie amount'] = df['star2 movie amount'].fillna(0)
df['star3 movie amount'] = df['star3 movie amount'].fillna(0)

df.head(20)

df.to_csv('to_ml.csv')


```

```python

numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
df = df.select_dtypes(include=numerics)
df = df.reset_index(drop=True)


df.head(10)



```

```python

# X = df[df.columns[(df.columns != 'rating') & (df.columns != 'year') & (df.columns != 'Unnamed: 0')]]

X = df[df.columns[(df.columns != 'rating')
                  &(df.columns != 'year') 
                  &(df.columns != 'idbm_year')]]
y = df['rating'].fillna(0)



```

```python
X.info()
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
model = sklearn.linear_model.LinearRegression().fit(X_train, y_train)
```

```python
newDf = pd.DataFrame({'rating actual': y_test,'rating predicted':model.predict(X_test)})
```

```python

newDf.to_csv('predicted.csv')

newDf['rating predicted'] = round(newDf['rating predicted'],1)

print(r2_score(y_test, newDf['rating predicted']))

newDf = newDf[newDf['rating actual'] != round(newDf['rating predicted'],1)]
```

```python
def check_star_rating(df, name):
    if df[df.star_1 == name]['star1 rating avg'].empty:
        if df[df.star_2 == name]['star2 rating avg'].empty:
            if df[df.star_3 == name]['star3 rating avg'].empty:
                return 7
            else:
                return df[df.star_3 == name]['star3 rating avg'].iloc[0]
        else:
            return df[df.star_2 == name]['star2 rating avg'].iloc[0] 
 
    else:
        return df[df.star_1 == name]['star1 rating avg'].iloc[0]
    
```
